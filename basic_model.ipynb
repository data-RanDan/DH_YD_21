```python
import requests 

def submit_results(predictions, submitter):
    jsonStr = {
        'submitter': submitter, 
        'predictions': base64.b64encode(gzip.compress(f'{predictions}'.encode())).decode()
    }
    
    url = "https://y-data-lb.dagsops.dagshub.com/y-data/api"
    headers = {'Content-Type': 'application/json'}

    resp = requests.post(url = url , headers = headers , json = jsonStr)
    return resp.text
```
